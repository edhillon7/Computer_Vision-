```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.zeros((200, 300), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (250, 150), 180, -1)  # Light gray rectangle
cv2.circle(img, (150, 100), 40, 255, -1)           # White circle inside
cv2.putText(img, 'IoT', (90, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (120), 2)
```

```python
thresh_value = 127  # You can adjust this threshold value
_, thresh_img = cv2.threshold(img, thresh_value, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Global Thresholding (T = {thresh_value})")
plt.imshow(thresh_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
cv2.imwrite("thresholded_output.jpg", thresh_img)
print("✅ Thresholded image saved as 'thresholded_output.jpg'")
```
